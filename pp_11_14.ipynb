```python
import pandas as pd
data = pd.read_csv('student_data.csv')
df = pd.DataFrame(data)

# 1. Count the number of students who have an HSC GPA greater than 3.5
hsc_gpa_count = df[df['hsc_gpa'] > 3.5].shape[0]
print(f"Number of students with HSC GPA greater than 3.5: {hsc_gpa_count}")
```

```python
chemical_eng_students = df[(df['department'] == "Chemical Engineering") & (df['cgpa'] > 3.0)]
print("\nStudents from 'Chemical Engineering' with CGPA above 3.0:")
print(chemical_eng_students)
```

```python
average_attendance_abroad = df[df['abroad_study']]['attendance'].mean()
print(f"\nAverage attendance for students who have studied abroad: {average_attendance_abroad}")
```

```python
average_hsc_gpa = df['hsc_gpa'].mean()
df['hsc_gpa'].fillna(average_hsc_gpa, inplace=True)
print("\nUpdated DataFrame after filling missing HSC GPA values:")
print(df)
```

```python
import matplotlib.pyplot as plt
students_per_department_year = df.groupby(['passing_year', 'department']).size().unstack(fill_value=0)
students_per_department_year.plot(kind='line', marker='o')
plt.title("Number of Students from Each Department Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Students")
plt.legend(title="Department")
plt.grid(True)
plt.show()
```
